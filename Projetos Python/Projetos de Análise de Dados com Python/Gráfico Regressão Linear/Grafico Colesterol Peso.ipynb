```python
import pandas as pd #Manipular tabelas numéricas e séries temporais.
import matplotlib.pyplot as plt #Criação de gráficos 2D e 3D com visualizações estáticas, animadas e interativas.
import seaborn as sns #Para a visualização de dados estatísticos.
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('dados_clientes_clusterizado.csv', sep=';')
```

```python
sns.set(style='darkgrid')    # Ajuste opcional para um fundo branco com linhas de grade
plt.figure(figsize=(15, 10))  # Ajuste opcional para o tamanho do gráfico
 
# Scatter Plot com legenda na parte superior direita e cores personalizadas
scatter_plot = sns.scatterplot(data=df, 
                               x='peso', 
                               y='colesterol', 
                               hue='nome_cluster', 
                               palette={'Risco Baixo': 'red', 
                                        'Risco Moderado': 'green', 
                                        'Risco Alto': 'blue'})
# Criando a linha de regressão linear
sns.regplot(data=df, x='peso', y='colesterol', scatter=False, color='black')
    
# Adicionando legenda na parte superior direita com fundo branco
scatter_plot.legend(loc='center right', facecolor='white')
 
# Nomeando o gráfico
plt.title('Distribuição de Clientes por Peso e Colesterol com Linha de Tendência')

# Exibindo o gráfico
plt.show()
```
