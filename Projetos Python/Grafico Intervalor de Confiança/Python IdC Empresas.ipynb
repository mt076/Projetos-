```python
import numpy as np 
import matplotlib.pyplot as plt 
import scipy.stats as stats
```

```python
#Exercício 2: A média das ações de empresas que compõem a Bovespa é de
#R$30 e o desvio padrão é de R$8,20. Suponha que os preços das ações se
#distribuam normalmente.
#(a) Qual a probabilidade de uma empresa ter um preço de, no mínimo, R$40 para
#suas ações?
#(b) Qual a probabilidade de uma empresa ter um preço não superior a R$20 para
#suas ações?
#(c) Qual o preço das ações para que a empresa seja incluída nas 10% maiores?
```

```python
mu = 30 
sigma = 8.20
```

```python
preco_a = 40 
z_a = (preco_a - mu) / sigma 
prob_a = 1 - stats.norm.cdf(z_a)
```

```python
x_vals = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000) 
y_vals = stats.norm.pdf(x_vals, mu, sigma)
```

```python
plt.figure(figsize=(10, 6)) 
plt.plot(x_vals, y_vals, label='Distribuição Normal') 
plt.fill_between(x_vals, 0, y_vals, where=(x_vals >= preco_a), color='skyblue', alpha=0.5, label=f'Área acima de R$40 = {prob_a:.4f}') 
plt.axvline(preco_a, color='red', linestyle='--', label='R$40') 
plt.title('Probabilidade de uma empresa ter um preço mínimo de R$40') 
plt.xlabel('Preço das ações (R$)') 
plt.ylabel('Densidade de Probabilidade') 
plt.legend() 
plt.grid(True) 
plt.show()
```

```python
preco_b = 20 
z_b = (preco_b - mu) / sigma 
prob_b = stats.norm.cdf(z_b)
```

```python
plt.figure(figsize=(10, 6)) 
plt.plot(x_vals, y_vals, label='Distribuição Normal') 
plt.fill_between(x_vals, 0, y_vals, where=(x_vals <= preco_b), color='skyblue', alpha=0.5, label=f'Área até R$20 = {prob_b:.4f}') 
plt.axvline(preco_b, color='red', linestyle='--', label='R$20') 
plt.title('Probabilidade de uma empresa ter um preço não superior a R$20') 
plt.xlabel('Preço das ações (R$)') 
plt.ylabel('Densidade de Probabilidade') 
plt.legend() 
plt.grid(True) 
plt.show()
```

```python
z_c = stats.norm.ppf(0.90) 
preco_c = mu + z_c * sigma
```

```python
plt.figure(figsize=(10, 6)) 
plt.plot(x_vals, y_vals, label='Distribuição Normal') 
plt.fill_between(x_vals, 0, y_vals, where=(x_vals >= preco_c), color='skyblue', alpha=0.5, label=f'Área acima de R${preco_c:.2f}') 
plt.axvline(preco_c, color='red', linestyle='--', label=f'R${preco_c:.2f}') 
plt.title('Preço das ações para estar nas 10% maiores') 
plt.xlabel('Preço das ações (R$)') 
plt.ylabel('Densidade de Probabilidade') 
plt.legend() 
plt.grid(True) 
plt.show()
```
