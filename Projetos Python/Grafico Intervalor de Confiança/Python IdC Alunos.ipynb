```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats 
```

```python
#a) Sorteando um aluno ao acaso, qual é a probabilidade que ele termine o exame antes de 100 minutos?

#Z = 100 - 120 / 15  = -20/15 = -1.33

#P(Z< -1.33) =  0.0918

#R: A probabilidade de terminar o exame antes dos 100 minutos é de 9.18%. 

#b) Qual deve ser o tempo de prova de modo a permitir que 95% dos vestibulandos terminem no prazo estipulado?

#Z = 0.95 ~ 1.645

#120 + 1.645 x 15 ~ 120 + 24.675 = 144.675

#R: 95% dos alunos podem terminar a prova em 144.68 minutos.

#c) Qual é o intervalo central de tempo, tal que 80% dos estudantes gastam para completar o exame?

#Z = 0.10 ~ -1.28
#Z = 0.90 ~  1.28

#120 - 1.28 x 15 = 120 - 19.2 = 100.8
#120 + 1.28 x 15 = 120 + 19.2 = 139.2

#R: O intervalo central é de 100.8 e 139.2 minutos.
```

```python
mu = 120
sigma = 15
x = 100

z = (x - mu) / sigma
probabilidade = stats.norm.cdf(z)

x_vals = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y_vals = stats.norm.pdf(x_vals, mu, sigma)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='Distribuição Normal')
plt.fill_between(x_vals, 0, y_vals, where=(x_vals <= x), color='skyblue', alpha=0.5, label=f'Área até 100 min = {probabilidade:.4f}')
plt.axvline(x, color='red', linestyle='--', label='100 minutos')
plt.title('Probabilidade de terminar o exame antes de 100 minutos')
plt.xlabel('Tempo (minutos)')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True)
plt.show()
```

```python
mu = 120
sigma = 15

# Calculando o valor z para 95%
z_95 = stats.norm.ppf(0.95)

# Convertendo para o tempo correspondente
tempo_95 = mu + z_95 * sigma

# Gráfico
x_vals = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y_vals = stats.norm.pdf(x_vals, mu, sigma)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='Distribuição Normal')
plt.fill_between(x_vals, 0, y_vals, where=(x_vals <= tempo_95), color='skyblue', alpha=0.5, label=f'Área até {tempo_95:.2f} min')
plt.axvline(tempo_95, color='red', linestyle='--', label=f'{tempo_95:.2f} minutos')
plt.title('Tempo de prova para 95% dos alunos terminarem')
plt.xlabel('Tempo (minutos)')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True)
plt.show()
```

```python
mu = 120
sigma = 15

# Calculando os valores z para 10% e 90%
z_10 = stats.norm.ppf(0.10)
z_90 = stats.norm.ppf(0.90)

# Convertendo para os tempos correspondentes
tempo_10 = mu + z_10 * sigma
tempo_90 = mu + z_90 * sigma

# Gráfico
x_vals = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y_vals = stats.norm.pdf(x_vals, mu, sigma)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='Distribuição Normal')
plt.fill_between(x_vals, 0, y_vals, where=(x_vals >= tempo_10) & (x_vals <= tempo_90), color='skyblue', alpha=0.5, label=f'80% entre {tempo_10:.2f} e {tempo_90:.2f} min')
plt.axvline(tempo_10, color='red', linestyle='--', label=f'{tempo_10:.2f} minutos')
plt.axvline(tempo_90, color='green', linestyle='--', label=f'{tempo_90:.2f} minutos')
plt.title('Intervalo de tempo central para 80% dos estudantes')
plt.xlabel('Tempo (minutos)')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True)
plt.show()
```
